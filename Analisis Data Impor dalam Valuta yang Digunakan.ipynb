## Deskripsi Data

Analisis data dilakukan dari dataset yang diperoleh dari situs data.go.id dibuat oleh Bank Indonesia (BI).
  
#### Nilai Impor Indonesia berdasarkan berdasarkan valuta asing
Dataset ini berisi mengenai nilai impor Indonesia berdasarkan valuta asing periode 2005-2013
 
Penjelasan mengenai Variabel pada Dataset ini:
a. tahun: Tahun
b. bulan: Bulan
c. kode_valuta: Kode Valuta
d. nama_valuta: Nama Valuta
e. nilai_impor: Nilai impor (dalam Ribu USD)

Catatan:
- Untuk "Impor yang tidak dapat diklasifikasikan" Terdiri dari barang yang diperoleh di pelabuhan oleh sarana pengangkut, barang untuk diperbaiki, dan penyesuaian cakupan ekspor industri Batam

```python
import pandas as pd
```

```python
impor_valuta = pd.read_csv('processed-nilai-impor-valuta-2005-2013.csv', delimiter = ',')
```

```python
impor_valuta     = impor_valuta[impor_valuta['nama_valuta'] != 'Impor yang tidak dapat diklasifikasikan  1)']
impor_valuta     = impor_valuta[impor_valuta['nama_valuta'] != 'Asuransi dan ongkos pengangkutan']
impor_valuta.head()
```

## Statistik Data

```python
print("total nilai impor 2005-2013 =", impor_valuta.sum()['nilai_impor'])
print("rata-rata nilai impor 2005-2013 =", impor_valuta.mean()['nilai_impor'])
print("maksimal nilai impor 2005-2013 =", impor_valuta.max()['nilai_impor'])
print("minimal nilai impor 2005-2013 =", impor_valuta.min()['nilai_impor'])
```

### Per negara

```python
#jumlah
impor_valuta.groupby('nama_valuta').sum()['nilai_impor']
```

```python
#rata-rata
impor_valuta.groupby('nama_valuta').mean()['nilai_impor']
```

```python
#maksimal
impor_valuta.groupby('nama_valuta').max()['nilai_impor']
```

```python
#minimal
impor_valuta.groupby('nama_valuta').min()['nilai_impor']
```

### Per tahun

```python
#jumlah
impor_valuta.groupby('tahun').sum()['nilai_impor']
```

```python
#rata-rata
impor_valuta.groupby('tahun').mean()['nilai_impor']
```

```python
#maksimal
impor_valuta.groupby('tahun').max()['nilai_impor']
```

```python
#minimal
impor_valuta.groupby('tahun').min()['nilai_impor']
```

# Visualisasi Data

### Agregat Per Tahun

```python
ax = impor_valuta.groupby('tahun').sum()['nilai_impor']
ax = ax.plot(kind='line', title ="Jumlah Nilai Impor Berdasarkan Valuta Per Tahun", legend=True, fontsize=20, figsize= (30,10))
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Jumlah", fontsize=12)
```

### Proporsi Per Valuta

```python
ax = impor_valuta.groupby('nama_valuta').sum()['nilai_impor']
ax = ax.plot(kind='pie', title ="Jumlah Nilai Impor Berdasarkan Valuta Per Tahun", legend=True, fontsize=24, figsize= (50,50))
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Jumlah", fontsize=12)
```
